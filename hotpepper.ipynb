```python
import copy 

lat = 35.7019863
lng = 139.7831018

params = {'lat': lat, 'lng': lng, 'format': 'json'}
response = requests.get(url, headers=headers, params=params)

list_num = 5
shop_details = {}
shop_list = {}
for i in range(list_num):
    name = response.json()['results']['shop'][i]['name']
    photo = response.json()['results']['shop'][i]['photo']['mobile']['l']
    address = response.json()['results']['shop'][i]['address']
    open_time = response.json()['results']['shop'][i]['open']
    logo_image = response.json()['results']['shop'][i]['logo_image']
    urls = response.json()['results']['shop'][i]['urls']['pc']
    
    shop_details['name'] = name
    shop_details['photo'] = photo
    shop_details['address'] = address
    shop_details['open_time'] = open_time
    shop_details['logo_image'] = logo_image
    shop_details['urls'] = urls
    
    shop_list[f'shop{i}'] = copy.copy(shop_details)
    
# shop_list

response.json()
```
